**üìå Suggested :**

In this case, although the percentage of missing values in the alcohol_intake feature is relatively high **(around 40%)**, the missing rate is not significantly different between individuals with heart disease (1) and those without it (0). This indicates that the missingness is not strongly dependent on the target variable and is likely occurring at random rather than following a systematic pattern.

Since dropping the rows would result in a substantial loss of data (over 20,000 records), it is not an ideal approach. Instead, a more effective strategy is to handle the missing values through imputation. One suitable method is to impute the missing entries based on the average alcohol intake within each heart disease class. This allows us to retain all records while ensuring that the imputed values are still relevant to the target distribution and preserve class-specific characteristics.

## **Data Preprocessing and Pipelines**

- **Split to Input Features and Target Columns**

```python
x = data.drop('heart_disease', axis=1)
y = data['heart_disease']
```

```python
x
```

```python
y
```

- **Split ( x and y ) to train and test**

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

#### **Numerical Data Preprocessing**

- **Select Numerical Columns**

```python
numeric_cols = x.select_dtypes(include=['number']).columns
numeric_cols
```

- **Missing Values ---> There is No Missing Values To Impute it**

- **Scaling**

```python
from sklearn.preprocessing import StandardScaler

R_Scaler = StandardScaler()

x_train[numeric_cols] = R_Scaler.fit_transform(x_train[numeric_cols])
x_test[numeric_cols] = R_Scaler.transform(x_test[numeric_cols])
```

#### **Categorical Data Preprocessing**

- **Select Categorical Columns**

```python
cate_cols = x.select_dtypes(include=['object']).columns
cate_cols
```

- **Impute Missing Values ( alcohol_intake )**

```python
from sklearn.impute import SimpleImputer

S_imputer = SimpleImputer(strategy='constant', fill_value='unknown')


x_train[['alcohol_intake']] = S_imputer.fit_transform(x_train[['alcohol_intake']])
x_test[['alcohol_intake']] = S_imputer.transform(x_test[['alcohol_intake']])
```

```python
x_train.alcohol_intake.isna().sum()
```

- **Encoding**

```python
for col in cate_cols :
    print(f'Column : {col}')
    print(x_train[col].nunique())
    print(x_train[col].unique())
    print("*" * 40)
```

- **Nominal Columns --> ( gender , smoking , alcohol_intake ) --> Using OneHotEncoder**

```python
ohe_cols = ['gender' , 'smoking' , 'alcohol_intake']
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')

x_train_ohe = pd.DataFrame(ohe.fit_transform(x_train[ohe_cols]), columns=ohe.get_feature_names_out(ohe_cols), index=x_train.index)
x_test_ohe = pd.DataFrame(ohe.transform(x_test[ohe_cols]), columns=ohe.get_feature_names_out(ohe_cols), index=x_test.index)
```

- **Ordinal Columns --> ( physical_activity , diet , stress_level ) --> Using OrdinalEncoder**
    - physical_activity ‚Üí [ Sedentary < Moderate < Active ]
    - diet ‚Üí [ Unhealthy < Average < Healthy ]
    - stress_level ‚Üí [ Low < Medium < High ]

```python
oe_cols = ['physical_activity' , 'diet' , 'stress_level']
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['Sedentary', 'Moderate', 'Active'],
                              ['Unhealthy', 'Average', 'Healthy'],
                                ['Low', 'Medium', 'High']])

x_train_oe = pd.DataFrame(oe.fit_transform(x_train[oe_cols]), columns=oe_cols, index=x_train.index)
x_test_oe = pd.DataFrame(oe.transform(x_test[oe_cols]), columns=oe_cols, index=x_test.index)
```

### **Concatenate Datasets**

```python
# reset index to avoid misalignment during concatenation

x_train.reset_index(drop=True, inplace=True)
x_test.reset_index(drop=True, inplace=True)
x_test_oe.reset_index(drop=True, inplace=True)
x_train_oe.reset_index(drop=True, inplace=True)
x_test_ohe.reset_index(drop=True, inplace=True)
x_train_ohe.reset_index(drop=True, inplace=True)
```

```python
x_train = pd.concat([x_train.drop(ohe_cols +  oe_cols , axis=1), x_train_ohe, x_train_oe], axis=1)
x_test = pd.concat([x_test.drop(ohe_cols +  oe_cols , axis=1), x_test_ohe, x_test_oe], axis=1)
```

```python
x_train
```

```python
y_train.reset_index(drop=True, inplace=True)
y_test.reset_index(drop=True, inplace=True)
```

## **Machine Learning**

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(max_iter=50000)

lr.fit(x_train, y_train)

print(f"Training Score : {lr.score(x_train, y_train)*100:.2f} %")
print(f"Testing Score  : {lr.score(x_test, y_test)*100:.2f} %")
```

```python
# decision_tree 
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(
    max_depth=10,
    random_state=42
)

dt.fit(x_train, y_train)

print(f"Training Score : {dt.score(x_train, y_train)*100:.2f}%")
print(f"Testing Score  : {dt.score(x_test, y_test)*100:.2f}%")
```

## **Pipeline**

- **Numerical Pipeline**

```python
from imblearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

num_pipeline = Pipeline(steps=[
    ('scaler', scaler)  
])

num_pipeline
```

- **Categorical Pipeline**

```python
cat1_cols = ['alcohol_intake']
cat1_cols
```

```python
# categorical pipeline 1

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

imputer = SimpleImputer(strategy='constant', fill_value='unknown')
ohe = OneHotEncoder(sparse_output=False, drop='first')

cat1_pipeline = Pipeline(steps=[
    ('imputer', imputer),
    ('ohe', ohe)
])

cat1_pipeline
```

```python
# cat2 Columns

cat2_cols = ['physical_activity' , 'diet' , 'stress_level']
```

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder(categories=[['Sedentary', 'Moderate', 'Active'],
                                     ['Unhealthy', 'Average', 'Healthy'],
                                     ['Low', 'Medium', 'High']

])

cat2_pipeline = Pipeline(steps=[
    ('encoder', encoder)
])

cat2_pipeline
```

```python
cat3_cols = ['gender' , 'smoking']
cat3_cols
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False, drop='first')

cat3_pipeline = Pipeline(steps=[
    ('ohe', ohe)
])

cat3_pipeline
```

##### **Column Transformer**

```python
from sklearn.compose import ColumnTransformer

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, numeric_cols),
        ('cat1', cat1_pipeline, cat1_cols),
        ('cat2', cat2_pipeline, cat2_cols),
        ('cat3', cat3_pipeline, cat3_cols)
    ]
)

preprocessor
```

##### **Final Model**

```python
lr_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor), 
    ('classifier', LogisticRegression(max_iter=50000))
])

lr_pipeline.fit(x, y)
```
